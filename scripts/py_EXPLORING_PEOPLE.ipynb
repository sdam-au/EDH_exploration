```python
people = []
for sublist in EDH_df["people"].tolist():
    people.extend(sublist)
```

```python
def count_people(row):
    try: 
        number = len(row["people"])
    except:
        number = 0
    return number

EDH_df["count_people"] = EDH_df.apply(lambda row: count_people(row), axis=1)
```

```python
def return_by_condition(sub_df, key, value):
    people = []
    for sublist in sub_df["people"].tolist():
        people.extend(sublist)
    filtered = []
    for hum in people:
        try: 
            key_value = hum[key]
        except: 
            key_value = ""
        if key_value not in value:
            filtered.append(hum)   
    return filtered
```

```python
epitaph_df = EDH_df[EDH_df["type_of_inscription_clean"]=="epitaph"]
```

```python
len(return_by_condition(epitaph_df, "gender", "female"))
```
